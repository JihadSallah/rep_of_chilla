## adding titles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool

#draw a lineplot
sns.lineplot(x="sepal_length",y="petal_width", data=phool)
plt.title("phoolon ka hisab kitab")
plt.show
```

## adding limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool

#draw a lineplot
sns.lineplot(x="sepal_length",y="petal_width", data=phool)
plt.title("phoolon ka hisab kitab")
plt.xlim(2)
plt.ylim(1)
plt.show
```

## Set Style

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)# we need to fix this error by adding object with function like sns.set_style
#load dataset
data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
sns.set_style("dark")
plt.title("Iris Line Plot") # add the title
plt.show()
```
