```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool=sns.load_dataset("iris")
phool

#draw a barplot
sns.barplot(x="species",y="petal_width", data=phool)
plt.title("phoolon ka hisab kitab")
#plt.xlim(2) #to put limit on ais
#plt.ylim(1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

 #load data set
boat=sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="sex",y="alone",hue="who",data=boat, order=["female","male"],color="red")
plt.show()
```

```python
#remove confidence interval

import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

 #load data set
boat=sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="sex",y="alone",hue="who",data=boat, order=["female","male"],color="red",ci=None,
           palette="pastel",estimator=mean)
plt.show()
```

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

 #load data set
boat=sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="class",y="fare",hue="sex", data=boat, estimator=mean,saturation=0.8)
sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.show()
```

```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

 #load data set
boat=sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="class",y="fare",hue="sex", data=boat,
           linewidth="3",facecolor=(1,0.3,0.3,0.4),
           edgecolor="0.5",errcolor="0.5") 
#(x, y, hue, data, order, hue_order, estimator, ci, n_boot, units, seed, orient,  
 # ...color, palette, saturation, errcolor, errwidth, capsize, dodge, ax, **kwargs)
plt.show()
```
